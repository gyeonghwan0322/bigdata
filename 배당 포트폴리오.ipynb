```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정 (예: 나눔고딕)
font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows 환경에서 '맑은 고딕' 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 사용자로부터 공백으로 구분된 숫자들을 입력받기
input_string = input("공백으로 구분된 숫자를 입력하세요: ")

# 입력된 문자열을 공백으로 나누어 숫자 리스트로 변환
numbers = input_string.split()

# 각 숫자의 등장 횟수를 세기
from collections import Counter
count = Counter(numbers)

# 중복 횟수 기준으로 내림차순 정렬
sorted_numbers = sorted(count.items(), key=lambda x: x[1], reverse=True)

# 각 숫자 뒤에 ".ks"를 붙여 리스트에 저장
result = [f"{number}.ks" for number, _ in sorted_numbers]

# MDD 계산 함수 정의
def calculate_mdd(data):
    running_max = data.cummax()
    drawdown = (data - running_max) / running_max
    mdd = drawdown.min()
    return mdd

# 종목코드별 MDD 계산
tickers = result
mdd_dict = {}

for ticker in tickers:
    try:
        stock_data = yf.download(ticker, period="3y", progress=False)['Adj Close']
        mdd = round(calculate_mdd(stock_data) * 100, 2)  # 퍼센트로 변환하고 소수 둘째자리에서 반올림
        mdd_dict[ticker] = mdd
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# MDD가 낮은 상위 10개 종목 선택
sorted_mdd = dict(sorted(mdd_dict.items(), key=lambda x: x[1]))
top_10_tickers = list(sorted_mdd.keys())[:10]

# 상위 10개 종목의 시가총액 가중 포트폴리오 구성
data_dict = {}
market_caps = []

for ticker in top_10_tickers:
    try:
        stock_data = yf.download(ticker, start="2021-06-01", end="2024-06-30", progress=False)
        data_dict[ticker] = stock_data['Adj Close']
        market_caps.append(stock_data['Adj Close'][-1] * stock_data['Volume'][-1])  # 시가총액 추정 (종가 * 거래량)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# 데이터프레임으로 변환
df_prices = pd.DataFrame(data_dict)

# 시가총액 가중치 계산
market_caps = np.array(market_caps)
weights = market_caps / market_caps.sum()

# 초기 투자금액 설정
total_investment = 1000000  # 1,000,000 원

# 각 종목에 투자된 금액 계산
initial_investments = total_investment * weights
initial_shares = initial_investments / df_prices.iloc[0]

# 포트폴리오의 일별 가치 계산
df_portfolio_value = (df_prices * initial_shares).sum(axis=1)

# 초기 및 마지막 투자금액
initial_value = df_portfolio_value.iloc[0]
final_value = df_portfolio_value.iloc[-1]

# 결과 출력
print("\n상위 10개 종목의 시가총액 가중 포트폴리오 누적 수익률:")
print(df_portfolio_value)
print(f"\n초기 투자금액: {initial_value:.2f} 원")
print(f"마지막 투자금액: {final_value:.2f} 원")

# 누적 수익률 그래프로 나타내기
plt.figure(figsize=(14, 7))
plt.plot(df_portfolio_value.index, df_portfolio_value, label='Portfolio Value', color='b')

plt.title('상위 10개 종목의 시가총액 가중 포트폴리오 누적 수익률 (2021년 6월 ~ 2024년 6월)')
plt.xlabel('Date')
plt.ylabel('Portfolio Value (KRW)')
plt.legend()  # 포트폴리오 레이블 포함
plt.grid()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

# 사용자로부터 공백으로 구분된 숫자들을 입력받기
input_string = input("공백으로 구분된 숫자를 입력하세요: ")

# 입력된 문자열을 공백으로 나누어 숫자 리스트로 변환
numbers = input_string.split()

# 각 숫자의 등장 횟수를 세기
from collections import Counter
count = Counter(numbers)

# 중복 횟수 기준으로 내림차순 정렬
sorted_numbers = sorted(count.items(), key=lambda x: x[1], reverse=True)

# 각 숫자 뒤에 ".ks"를 붙여 리스트에 저장
result = [f"{number}.ks" for number, _ in sorted_numbers]

# MDD 계산 함수 정의
def calculate_mdd(data):
    running_max = data.cummax()
    drawdown = (data - running_max) / running_max
    mdd = drawdown.min()
    return mdd

# 종목코드별 MDD, 수익률, 변동성 계산
tickers = result
summary_data = []

for ticker in tickers:
    try:
        stock_data = yf.download(ticker, start="2021-06-01", end="2024-06-30", progress=False)['Adj Close']
        mdd = round(calculate_mdd(stock_data) * 100, 2)  # 퍼센트로 변환하고 소수 둘째자리에서 반올림
        total_return = round(((stock_data.iloc[-1] / stock_data.iloc[0]) - 1) * 100, 2)  # 총 수익률 계산
        volatility = round(stock_data.pct_change().std() * np.sqrt(252) * 100, 2)  # 연간 변동성 계산
        summary_data.append({'Ticker': ticker, 'MDD (%)': mdd, 'Total Return (%)': total_return, 'Volatility (%)': volatility})
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# 데이터프레임으로 정리
df_summary = pd.DataFrame(summary_data)

# 결과 출력
print("\n종목코드별 MDD, 수익률, 변동성:")
print(df_summary)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글 폰트 설정 (예: 나눔고딕)
font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows 환경에서 '맑은 고딕' 폰트 경로
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 종목코드, 회사명, 평균 배당수익률 데이터 정의
data_info = [
    {"ticker": "029780.ks", "name": "삼성카드", "dividend_yield": 7.3},
    {"ticker": "001720.ks", "name": "신영증권", "dividend_yield": 6.8},
    {"ticker": "033780.ks", "name": "케이티앤지", "dividend_yield": 6.0},
    {"ticker": "002960.ks", "name": "한국쉘석유", "dividend_yield": 7.7},
    {"ticker": "015360.ks", "name": "예스코홀딩스", "dividend_yield": 7.0},
    {"ticker": "003460.ks", "name": "유화증권", "dividend_yield": 4.9},
    {"ticker": "030200.ks", "name": "KT", "dividend_yield": 5.7},
    {"ticker": "024110.ks", "name": "기업은행", "dividend_yield": 8.3},
    {"ticker": "175330.ks", "name": "JB금융지주", "dividend_yield": 7.5},
    {"ticker": "000810.ks", "name": "삼성화재", "dividend_yield": 6.08}
]

# 동일가중 포트폴리오 구성 및 배당금 포함 데이터 수집
data_dict = {}
dividend_yields = {}

for data in data_info:
    ticker = data['ticker']
    dividend_yield = data['dividend_yield']
    try:
        stock_data = yf.download(ticker, start="2021-06-01", end="2024-06-30", progress=False)
        data_dict[ticker] = stock_data['Adj Close']
        dividend_yields[ticker] = dividend_yield / 100  # 배당수익률을 백분율로 변환
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# 데이터프레임으로 변환
df_prices = pd.DataFrame(data_dict)

# 동일 가중 포트폴리오 구성 (각 종목에 동일한 금액 투자)
initial_investment = 1000000  # 총 투자금액 1,000,000 원
weight = 1 / len(data_info)  # 동일 가중
initial_investments = initial_investment * weight
initial_shares = initial_investments / df_prices.iloc[0]

# 포트폴리오의 일별 가치 계산 (배당금 포함)
df_dividends = pd.DataFrame({ticker: df_prices[ticker] * initial_shares[ticker] * dividend_yields[ticker] / 252 for ticker in df_prices.columns})
df_total_value = (df_prices * initial_shares).sum(axis=1) + df_dividends.sum(axis=1).cumsum()

# 시가총액 가중 포트폴리오 구성
data_market_cap = {}
market_caps = []

for data in data_info:
    ticker = data['ticker']
    try:
        stock_data = yf.download(ticker, start="2021-06-01", end="2024-06-30", progress=False)
        data_market_cap[ticker] = stock_data['Adj Close']
        market_caps.append(stock_data['Adj Close'][-1] * stock_data['Volume'][-1])  # 시가총액 추정 (종가 * 거래량)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

# 데이터프레임으로 변환
df_market_cap_prices = pd.DataFrame(data_market_cap)

# 시가총액 가중치 계산
market_caps = np.array(market_caps)
weights = market_caps / market_caps.sum()
initial_investments_market_cap = initial_investment * weights
initial_shares_market_cap = initial_investments_market_cap / df_market_cap_prices.iloc[0]

# 시가총액 가중 포트폴리오의 일별 가치 계산 (배당금 포함)
df_market_cap_dividends = pd.DataFrame({ticker: df_market_cap_prices[ticker] * initial_shares_market_cap[ticker] * dividend_yields[ticker] / 252 for ticker in df_market_cap_prices.columns})
df_market_cap_total_value = (df_market_cap_prices * initial_shares_market_cap).sum(axis=1) + df_market_cap_dividends.sum(axis=1).cumsum()

# 결과 출력
initial_value_equal = df_total_value.iloc[0]
final_value_equal = df_total_value.iloc[-1]
initial_value_market_cap = df_market_cap_total_value.iloc[0]
final_value_market_cap = df_market_cap_total_value.iloc[-1]

total_dividends_equal = df_dividends.sum(axis=1).cumsum().iloc[-1]
total_dividends_market_cap = df_market_cap_dividends.sum(axis=1).cumsum().iloc[-1]

print("\n동일 가중 포트폴리오 누적 수익률 (배당금 포함):")
print(df_total_value)
print(f"\n초기 투자금액 (동일가중): {initial_value_equal:.2f} 원")
print(f"마지막 투자금액 (동일가중): {final_value_equal:.2f} 원")
print(f"총 배당금 (동일가중): {total_dividends_equal:.2f} 원")

print("\n시가총액 가중 포트폴리오 누적 수익률 (배당금 포함):")
print(df_market_cap_total_value)
print(f"\n초기 투자금액 (시가총액가중): {initial_value_market_cap:.2f} 원")
print(f"마지막 투자금액 (시가총액가중): {final_value_market_cap:.2f} 원")
print(f"총 배당금 (시가총액가중): {total_dividends_market_cap:.2f} 원")

# 누적 수익률 그래프 출력
plt.figure(figsize=(14, 7))
plt.plot(df_total_value.index, df_total_value, label='Equal Weighted Portfolio Value', color='b')
plt.plot(df_market_cap_total_value.index, df_market_cap_total_value, label='Market Cap Weighted Portfolio Value', color='g')
plt.title('포트폴리오 누적 수익률 (배당금 포함) (2021년 6월 ~ 2024년 6월)')
plt.xlabel('Date')
plt.ylabel('Portfolio Value (KRW)')
plt.legend()  # 포트폴리오 레이블 포함
plt.grid()
plt.show()

# 원형 포트폴리오 차트 출력
plt.figure(figsize=(10, 7))
plt.pie([weight] * len(data_info), labels=[data['name'] for data in data_info], autopct='%1.1f%%', startangle=140)
plt.title('동일 가중 포트폴리오 구성')
plt.show()

plt.figure(figsize=(10, 7))
plt.pie(weights, labels=[data['name'] for data in data_info], autopct='%1.1f%%', startangle=140)
plt.title('시가총액 가중 포트폴리오 구성')
plt.show()
```
