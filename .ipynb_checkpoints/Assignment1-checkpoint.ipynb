```python
import pandas as pd
import numpy as np
import itertools
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter("ignore", UserWarning)

# 필요한 라이브러리 임포트
# ...

# 데이터 불러오기 및 전처리 (이전 코드와 동일)
df = pd.read_excel("Welch_Goyal_PredictorData2023.xlsx", sheet_name='Quarterly')

# 프리미엄 및 예측 변수 생성
df['premium'] = df['CRSP_SPvw'] - df['Rfree']
df['dp'] = np.log(df['D12']) - np.log(df['Index'])
df['dy'] = np.log(df['D12']) - np.log(df['Index'].shift(1))
df['ep'] = np.log(df['E12']) - np.log(df['Index'])
df['tms'] = df['lty'] - df['tbl']
df['dfy'] = df['BAA'] - df['AAA']
df = df.rename(columns={'b/m':'bm'})
x_names = ['dp', 'dy', 'ep','bm', 'ntis', 'tbl', 'ltr', 'tms', 'dfy', 'infl', 'ik']

# 날짜 컬럼 생성
df['year'] = df['yyyyq'].astype(str).str[:4].astype(int)
df['quarter'] = df['yyyyq'].astype(str).str[4].astype(int)
df['date'] = pd.PeriodIndex(year=df['year'], quarter=df['quarter'], freq='Q')
```

```python

# 예측 변수 시차 적용 및 결측치 제거
df_clean = pd.concat([df[['date', 'premium']], df[x_names].shift(1)], axis=1 )
df_clean = df_clean.dropna(axis=0)
df_clean
```

```python

# 예측 결과를 저장할 데이터프레임 생성
results = df_clean[['date', 'premium']].copy()
results = results[results['date'] >= '1970Q1']  # 1970Q1부터 예측 시작
results = results.reset_index(drop=True)
```

```python
# 예측값을 저장할 리스트 초기화
forecast_values = []
benchmark_forecast_values = []

# 모든 시점에 대해 재귀적으로 모델 추정 및 예측
for i in range(len(results)):
    forecast_date = results.loc[i, 'date']
    # 훈련 데이터 설정: 현재 예측 시점 이전의 모든 데이터
    train_data = df_clean[df_clean['date'] < forecast_date]
    X_train = train_data[x_names]
    y_train = train_data['premium']
    
    # 테스트 데이터 설정: 현재 예측 시점의 데이터
    X_test = df_clean[df_clean['date'] == forecast_date][x_names]
    y_test = df_clean[df_clean['date'] == forecast_date]['premium'].values[0]
    
    # 상수항 추가
    X_train_const = sm.add_constant(X_train)
    X_test_const = X_test.insert(0, 'const' , 1)
    
    # OLS 회귀 모델 추정
    model = sm.OLS(y_train, X_train_const).fit()
    
    # 예측 수행
    forecast = model.predict(X_test_const)
    forecast_values.append(forecast[0])
    
    # 벤치마크 모델 예측: 훈련 데이터의 프리미엄 평균 사용
    mean_premium = y_train.mean()
    benchmark_forecast_values.append(mean_premium)
    
    # 진행 상황 출력 (선택 사항)
    print(f"Forecast for {forecast_date}: {forecast[0]:.4f}, Actual: {y_test:.4f}")

# 결과 데이터프레임에 예측값 추가
results['Forecast'] = forecast_values
results['Benchmark'] = benchmark_forecast_values

# 실제값과 예측값 비교
print(results.head())

# 외부표본 MSE 계산
mse_model = np.mean((results['premium'] - results['Forecast'])**2)
mse_benchmark = np.mean((results['premium'] - results['Benchmark'])**2)

print(f"Out-of-sample MSE of the model: {mse_model:.6f}")
print(f"Out-of-sample MSE of the benchmark model: {mse_benchmark:.6f}")

# 예측값과 실제값, 벤치마크 모델 예측값을 함께 플롯
plt.figure(figsize=(14, 7))
plt.plot(results['date'].dt.to_timestamp(), results['premium'], label='Actual', marker='o')
plt.plot(results['date'].dt.to_timestamp(), results['Forecast'], label='Forecast', marker='o')
plt.plot(results['date'].dt.to_timestamp(), results['Benchmark'], label='Benchmark', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Premium')
plt.title('Actual vs. Forecasted Stock Returns')
plt.legend()
plt.show()
```

```python
current_date = '2023Q3'
forecast_date = '2023Q4'

#2023년 3분기까지의 데이터 중 x_names의 리스트에 포함된 모든 변수를 포함한 훈련 데이터의 특성행렬을 생성
X_train = df_clean.loc[df_clean['date'] <= current_date, x_names]
y_train = df_clean.loc[df_clean['date'] <= current_date, 'premium']

#2023년 4분기의 test데이터 생성
X_test = df_clean.loc[df_clean['date'] == forecast_date, x_names]
y_test = df_clean.loc[df_clean['date'] == forecast_date, 'premium']
```

```python
#(a) AIC를 사용한 모델 선택
# AIC 계산 함수
def calculate_aic(n, resid, k):
    sse = np.sum(resid ** 2)
    aic = n * np.log(sse / n) + 2 * k
    return aic

# 가능한 모든 모델에 대해 AIC 계산
aic_values = []
models = []
n = len(y_train)

for i in range(1, len(x_names) + 1):
    for combo in itertools.combinations(x_names, i):
        X_train_selected = sm.add_constant(X_train[list(combo)], has_constant='add')
        model = sm.OLS(y_train, X_train_selected).fit()
        resid = model.resid
        k = len(combo) + 1  # 상수항 포함
        aic = calculate_aic(n, resid, k)
        aic_values.append((aic, combo))
        models.append((model, combo))

# AIC가 가장 작은 모델 선택
best_aic, best_combo = min(aic_values, key=lambda x: x[0])
best_model_index = aic_values.index((best_aic, best_combo))
best_model = models[best_model_index][0]

print(f"AIC가 가장 작은 모델의 예측 변수: {best_combo}")

# 2023Q4 예측
X_test_selected = X_test[list(best_combo)].copy()
X_test_selected = sm.add_constant(X_test_selected, has_constant='add')
forecast = best_model.predict(X_test_selected)

print(f"2023Q4 주식 수익률 예측 (AIC 기준): {forecast.iloc[0]:.4f}")
# 실제 y_test 값이 있는 경우 출력
if not y_test.empty:
    print(f"실제 2023Q4 주식 수익률: {y_test.iloc[0]:.4f}")
```

```python
# BIC 계산 함수
def calculate_bic(n, resid, k):
    sse = np.sum(resid ** 2)
    bic = n * np.log(sse / n) + k * np.log(n)
    return bic

# 가능한 모든 모델에 대해 BIC 계산
bic_values = []
models = []
n = len(y_train)

for i in range(1, len(x_names) + 1):
    for combo in itertools.combinations(x_names, i):
        X_train_selected = sm.add_constant(X_train[list(combo)], has_constant='add')
        model = sm.OLS(y_train, X_train_selected).fit()
        resid = model.resid
        k = len(combo) + 1  # 상수항 포함
        bic = calculate_bic(n, resid, k)
        bic_values.append((bic, combo))
        models.append((model, combo))

# BIC가 가장 작은 모델 선택
best_bic, best_combo = min(bic_values, key=lambda x: x[0])
best_model_index = bic_values.index((best_bic, best_combo))
best_model = models[best_model_index][0]

print(f"BIC가 가장 작은 모델의 예측 변수: {best_combo}")

# 2023Q4 예측
X_test_selected = X_test[list(best_combo)].copy()
X_test_selected = sm.add_constant(X_test_selected, has_constant='add')
forecast = best_model.predict(X_test_selected)

print(f"2023Q4 주식 수익률 예측 (BIC 기준): {forecast.iloc[0]:.4f}")
if not y_test.empty:
    print(f"실제 2023Q4 주식 수익률: {y_test.iloc[0]:.4f}")
```

```python
# v 계산
t = len(y_train)
v = int(0.25 * (0.5 * t - np.sqrt(t - 1)))
if v < 1:
    v = 1  # 최소값 보정
Tv = 2 * v + 1

# 가능한 모든 모델에 대해 v-블록 교차 검증 수행
cv_errors = []
models = []

for i in range(1, len(x_names) + 1):
    for combo in itertools.combinations(x_names, i):
        errors = []
        # v-블록 교차 검증
        indices = np.arange(t)
        np.random.seed(0)  # 재현성을 위해 시드 설정
        np.random.shuffle(indices)
        fold_sizes = np.full(v, t // v)
        fold_sizes[:t % v] += 1
        current = 0
        for fold_size in fold_sizes:
            test_indices = indices[current:current + fold_size]
            train_indices = np.setdiff1d(indices, test_indices)
            current += fold_size
            
            X_cv_train = X_train.iloc[train_indices]
            y_cv_train = y_train.iloc[train_indices]
            X_cv_test = X_train.iloc[test_indices]
            y_cv_test = y_train.iloc[test_indices]
            
            # 선택된 변수로 모델 적합
            X_cv_train_selected = sm.add_constant(X_cv_train[list(combo)], has_constant='add')
            model = sm.OLS(y_cv_train, X_cv_train_selected).fit()
            # 검증 세트에 대한 예측
            X_cv_test_selected = sm.add_constant(X_cv_test[list(combo)], has_constant='add')
            y_pred = model.predict(X_cv_test_selected)
            # 오류 계산
            error = mean_squared_error(y_cv_test, y_pred)
            errors.append(error)
        # 평균 오류 계산
        mean_error = np.mean(errors)
        cv_errors.append((mean_error, combo))
        models.append((model, combo))

# 검증 오류가 가장 작은 모델 선택
best_cv_error, best_combo = min(cv_errors, key=lambda x: x[0])
best_model_index = cv_errors.index((best_cv_error, best_combo))
best_model = models[best_model_index][0]

print(f"v-블록 교차 검증으로 선택된 예측 변수: {best_combo}")

# 2023Q4 예측
X_test_selected = X_test[list(best_combo)].copy()
X_test_selected = sm.add_constant(X_test_selected, has_constant='add')
forecast = best_model.predict(X_test_selected)

print(f"2023Q4 주식 수익률 예측 (v-블록 CV 기준): {forecast.iloc[0]:.4f}")
if not y_test.empty:
    print(f"실제 2023Q4 주식 수익률: {y_test.iloc[0]:.4f}")
```

```python
# 전진 선택법 함수 정의
def forward_stepwise(X, y, entry_threshold=0.05):
    initial_features = []
    remaining_features = list(X.columns)
    selected_features = initial_features.copy()
    while remaining_features:
        p_values = pd.Series(index=remaining_features, dtype=float)
        for feature in remaining_features:
            X_selected = X[selected_features + [feature]]
            X_selected = sm.add_constant(X_selected, has_constant='add')
            model = sm.OLS(y, X_selected).fit()
            p_values[feature] = model.pvalues[feature]
        min_p_value = p_values.min()
        if min_p_value < entry_threshold:
            best_feature = p_values.idxmin()
            selected_features.append(best_feature)
            remaining_features.remove(best_feature)
        else:
            break
    return selected_features

# 전진 선택법 적용
selected_features = forward_stepwise(X_train, y_train, entry_threshold=0.05)
print(f"전진 선택법으로 선택된 예측 변수: {selected_features}")

# 선택된 변수로 모델 적합 및 예측
if selected_features:
    X_train_selected = X_train[selected_features]
    X_test_selected = X_test[selected_features]
    
    X_train_selected = sm.add_constant(X_train_selected, has_constant='add')
    X_test_selected = sm.add_constant(X_test_selected, has_constant='add')
    
    model = sm.OLS(y_train, X_train_selected).fit()
    forecast = model.predict(X_test_selected)
    
    print(f"2023Q4 주식 수익률 예측 (전진 선택법 기준): {forecast.iloc[0]:.4f}")
    if not y_test.empty:
        print(f"실제 2023Q4 주식 수익률: {y_test.iloc[0]:.4f}")
else:
    print("전진 선택법에 의해 선택된 예측 변수가 없습니다.")
```

```python
# AIC에 의해 선택된 예측 변수들
selected_predictors = ['dy', 'tbl', 'ltr', 'ik']
# 예측 변수 시차 적용 및 결측치 제거
df_clean = pd.concat([df[['date', 'premium']], df[selected_predictors].shift(1)], axis=1)
df_clean = df_clean.dropna(axis=0)

# 1990Q1부터 2023Q4까지의 예측 생성
start_date = '1990Q1'
end_date = '2023Q4'
forecast_dates = df_clean[(df_clean['date'] >= start_date) & (df_clean['date'] <= end_date)]['date'].unique()

# 결과를 저장할 리스트 초기화
forecasts = []
benchmarks = []
actuals = []
dates = []

for forecast_date in forecast_dates:
    # 훈련 데이터 설정
    train_data = df_clean[df_clean['date'] < forecast_date]
    X_train = train_data[selected_predictors]
    y_train = train_data['premium']
    
    # 테스트 데이터 설정
    test_data = df_clean[df_clean['date'] == forecast_date]
    X_test = test_data[selected_predictors]
    y_test = test_data['premium'].values[0]
    
    t = len(y_train)
    v = int(0.25 * (0.5 * t - np.sqrt(t - 1)))
    if v < 1:
        v = 1  # 최소값 보정
    Tv = 2 * v + 1
    
    # 가능한 모든 예측 변수 조합 생성
    predictors_combinations = []
    for i in range(1, len(selected_predictors) + 1):
        predictors_combinations.extend(itertools.combinations(selected_predictors, i))
    
    # v-블록 교차 검증을 사용하여 최적의 모델 선택
    cv_errors = []
    models = []
    n = len(y_train)
    
    for combo in predictors_combinations:
        errors = []
        indices = np.arange(n)
        fold_sizes = np.full(v, n // v)
        fold_sizes[:n % v] += 1
        current = 0
        for fold_size in fold_sizes:
            test_indices = indices[current:current + fold_size]
            train_indices = np.setdiff1d(indices, test_indices)
            current += fold_size
            
            X_cv_train = X_train.iloc[train_indices]
            y_cv_train = y_train.iloc[train_indices]
            X_cv_test = X_train.iloc[test_indices]
            y_cv_test = y_train.iloc[test_indices]
            
            # 선택된 변수로 모델 적합
            X_cv_train_const = sm.add_constant(X_cv_train[list(combo)], has_constant='add')
            model = sm.OLS(y_cv_train, X_cv_train_const).fit()
            
            # 검증 세트에 대한 예측
            X_cv_test_const = sm.add_constant(X_cv_test[list(combo)], has_constant='add')
            y_pred = model.predict(X_cv_test_const)
            
            # 오류 계산
            error = mean_squared_error(y_cv_test, y_pred)
            errors.append(error)
        # 평균 오류 계산
        mean_error = np.mean(errors)
        cv_errors.append((mean_error, combo))
        models.append((model, combo))
    
    # 검증 오류가 가장 작은 모델 선택
    best_cv_error, best_combo = min(cv_errors, key=lambda x: x[0])
    best_model_index = cv_errors.index((best_cv_error, best_combo))
    best_model = models[best_model_index][0]
    
    # 테스트 데이터에 대한 예측
    X_test_const = sm.add_constant(X_test[list(best_combo)], has_constant='add')
    forecast = best_model.predict(X_test_const)
    
    # 벤치마크 모델 예측 (훈련 데이터 평균)
    benchmark = y_train.mean()
    
    # 결과 저장
    forecasts.append(forecast.values[0])
    benchmarks.append(benchmark)
    actuals.append(y_test)
    dates.append(forecast_date)
    
    # 진행 상황 출력
    print(f"Date: {forecast_date}, Forecast: {forecast.values[0]:.4f}, Actual: {y_test:.4f}, Benchmark: {benchmark:.4f}")

# 결과를 데이터프레임으로 저장
results = pd.DataFrame({
    'Date': dates,
    'Actual': actuals,
    'Forecast': forecasts,
    'Benchmark': benchmarks
})

# MSE 계산
mse_model = mean_squared_error(results['Actual'], results['Forecast'])
mse_benchmark = mean_squared_error(results['Actual'], results['Benchmark'])

print(f"\n외부표본 MSE (모델): {mse_model:.6f}")
print(f"외부표본 MSE (벤치마크): {mse_benchmark:.6f}")

# 결과 플롯
plt.figure(figsize=(14, 7))
plt.plot(results['Date'].astype(str), results['Actual'], label='Actual', marker='o')
plt.plot(results['Date'].astype(str), results['Forecast'], label='Forecast', marker='o')
plt.plot(results['Date'].astype(str), results['Benchmark'], label='Benchmark', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Stock Return')
plt.title('Actual vs Forecasted Stock Returns (1990Q1 - 2023Q4)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
